```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Street_Tree_List-2022-01-30_FILTERED_v0.csv")
```

```python
df['Genus'] = df['qSpecies'].str.split().str[0]
df['ScientificName'] = df['qSpecies'].str.split().str[0:2].str.join(' ')
df['CommonName'] = df['qSpecies'].str.split('::').str[1].str.strip()

def process_value(row):
    row_str = str(row)
    parts = row_str.split('/') 
    if len(parts) > 2:  
        num_str = parts[2]  
        try:
            num = int(num_str)  
            if num > 24:
                return "19" + num_str
            else:
                return "20" + num_str
        except ValueError:
            return None  
    else:
        return None  

df['Year'] = df['PlotSize'].apply(process_value)
```

```python
type(df["PlantDate"])
```

```python
df['Date'] = pd.to_datetime(df['PlantDate'])
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year.astype('Int64') 
df["Age"] = 2024-df['Year']
```

```python
df_age = df.dropna(subset=['Age'])
df_age
```

```python
df_year_sum = df_age.groupby('Year').size().reset_index(name='count')
df_year_sum = df_year_sum.rename(columns={'Year': 'year'})
df_year_sum
```

```python
popular_trees = [
    "Sycamore: London Plane",
    "New Zealand Xmas Tree",
    "Brisbane Box",
    "Victorian Box",
    "Swamp Myrtle"
]

popular_df = df_age[df_age["CommonName"].isin(popular_trees)]
sorted_filtered_df = popular_df.sort_values(by="Year")
tree_counts = popular_df["CommonName"].value_counts()
tree_counts
```

```python
filtered_df = df_age[df_age["CommonName"].isin(popular_trees)]
grouped_df = filtered_df.groupby(["CommonName", "Year"]).size().reset_index(name="count")
grouped_df = grouped_df.rename(columns={"CommonName": "species"})
grouped_df = grouped_df.rename(columns={"Year": "year"})
grouped_df
```
